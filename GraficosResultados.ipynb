```python
from turtle import color


algorithms = ["E2V2D1", "E2V2D1C1", "MOEAD", "NSGA2", "NSGA3"]
DTLZ_n = [1, 2, 3, 5, 6, 7]

fig, axs = plt.subplots(len(DTLZ_n),len(algorithms),figsize=(30,30), subplot_kw=dict(projection='3d'))

for i, n in enumerate(DTLZ_n):
    for j, alg in enumerate(algorithms):
        with open("result/"+alg+"_DTLZ"+str(n)+"_3obj.pkl", "rb") as f:
                results = pickle.load(f)['combined'][1]
                
        hv = get_performance_indicator("hv", ref_point=get_problem("wfg"+str(n), n_var=10, n_obj=3).nadir_point())
        
        axs[i,j].scatter(results[:, 0], results[:, 1], results[:, 2])
        position = axs[i,j].get_position()
        fig.text(position.bounds[0] + 0.035, position.bounds[1] - 0.013, hv.do(results), ha='center', va='center', color='red', fontsize=10)
        
        if(i == 0):
            axs[i,j].set_title(alg, fontsize=30)
        if(j == 0):
            fig.text(position.bounds[0] - 0.02, position.bounds[1] + 0.03, "DTLZ"+str(n), ha='center', va='center', rotation='vertical',fontsize=30)


plt.show()
```
